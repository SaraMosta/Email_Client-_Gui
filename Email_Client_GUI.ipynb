```python
import tkinter as tk
import smtplib
from tkinter import messagebox
```

```python
def sendemail():
    sub = subjectEntry.get()
    msg = text_area.get("1.0", "end-1c")  # Get the message from Text widget
    From = fromEntryField.get()
    to = toEntryField.get()
    password=passEntry.get()

    # create smtp server
    server = smtplib.SMTP('smtp.work.com', 587)
    server.ehlo()

    # auth
    server.login(From, password)

    message = 'Subject:' + sub + '\n' + msg

    server.sendmail(From, to,  f'Subject: {sub}\n\n{msg}')
    
    server.close()

    # Display success message
    messagebox.showinfo("Success", "Email sent successfully!")
```

```python
root = tk.Tk()
root.title('SEND EMAIL GUI')
root.geometry('820x900')
root.config(background='#79a6d2')
root.resizable(False, False)

Frame_Title = tk.Frame(root, bg='grey')
Frame_Title.grid(row=0, column=0)


from_label = tk.Label(root, text="From:",font=(
    'times new roman', 20, 'bold'), bd=5, fg='white', bg='#79a6d2')
from_label.place(x=30, y=18)


fromLabelFrame = tk.Label(root,font=(
    'times new roman', 10, 'bold'), bd=5, fg='white', bg='#3973ac')
fromLabelFrame.place(x=160, y=22)

fromEntryField = tk.Entry(fromLabelFrame, font=(
    'times new roman', 18), width=50)
fromEntryField.grid(row=0, column=0)


to_label = tk.Label(root, text="To:",font=(
    'times new roman', 20, 'bold'), bd=5, fg='white', bg='#79a6d2')
to_label.place(x=40, y=100)
toLabelFrame = tk.LabelFrame(root,font=(
    'times new roman', 10, 'bold'), bd=5, fg='white', bg='#4080bf')
toLabelFrame.place(x=160, y=100)

toEntryField = tk.Entry(toLabelFrame, font=(
    'times new roman', 18), width=50)
toEntryField.grid(row=0, column=0)


subject_label = tk.Label(root, text="Subject:",font=(
    'times new roman', 20, 'bold'), bd=5, fg='white', bg='#79a6d2')
subject_label.place(x=20, y=180)

subjectLabelFrame = tk.Label(root,font=(
    'times new roman', 10, 'bold'), bd=5, fg='white', bg='#3973ac')
subjectLabelFrame.place(x=160, y=180)

subjectEntry = tk.Entry(subjectLabelFrame, font=(
    'times new roman', 18), width=50)
subjectEntry.grid(row=0, column=0)


message_label = tk.Label(root, text="Message:",font=(
    'times new roman', 20, 'bold'), bd=5, fg='white', bg='#79a6d2')
message_label.place(x=20, y=350)
emailLabelFrame = tk.LabelFrame(root,font=(
    'times new roman', 10, 'bold'), bd=5, fg='white', bg='#4080bf')
emailLabelFrame.place(x=160, y=260)

text_area = tk.Text(emailLabelFrame, font=('times new roman', 14,), height=10,width=67)
text_area.grid(row=1, column=0, columnspan=2)


password_label = tk.Label(root, text="Password:",font=(
    'times new roman', 20, 'bold'), bd=5, fg='white', bg='#79a6d2')
password_label.place(x=20, y=520)

PassLabelFrame = tk.LabelFrame(root,font=(
    'times new roman', 10, 'bold'), bd=5, fg='white', bg='#4080bf')
PassLabelFrame.place(x=160, y=523)

passEntry = tk.Entry(PassLabelFrame,show="*", font=(
    'times new roman', 18), width=50)
passEntry.grid(row=0, column=0)

sendbutton = tk.Button(root, text='Send', bd=0, bg='#204060', fg='white', font='lucida 17 bold',
                       cursor='hand2', command=sendemail, width=30)
sendbutton.place(x=250, y=600)

successLabel = tk.LabelFrame(root,text='',font=(
    'times new roman', 10, 'bold'), bd=5, fg='#204060', bg='#4080bf')
successLabel.place(x=10, y=670)



root.mainloop()
```
